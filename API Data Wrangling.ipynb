These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.

```python
# 1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
url1 = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&api_key='+API_KEY
r = requests.get(url1)
file = r.text
print(file[0:3000])
```

```python
# 2. Convert the returned JSON object into a Python dictionary.data = json.loads(file)
data = json.loads(file)
data = data.get('dataset')
```

```python
col_names = data.get('column_names')
print(col_names)
```

```python
# 3. Calculate what the highest and lowest opening prices were for the stock in this period.
data_list= data.get('data')
high_open = 0
for element in data_list:
    if type(element[1]) != float:
        continue
    if high_open <= element[1]:
        high_open = element[1]   
print('The highest opening price is '+ str(high_open))

low_open = high_open
for element in data_list:
    if type(element[1]) != float:
        continue
    if low_open > element[1]:
        low_open = element[1]   
    
print('The lowest opening price is '+ str(low_open))
```

```python
# 4. What was the largest change in any one day (based on High and Low price)?
change = 0
for element in data_list:
    if type(element[2]) != float:
        continue
    if type(element[3]) != float:
        continue
    if element[2]-element[3] > change:
        change = element[2]-element[3]  
change = round(change,2)
print('The largest price change in a day is '+ str(change))
```

```python
# 5. What was the largest change between any two days (based on Closing Price)?
change = 0
for i in range(len(data_list)-1):
    if type(data_list[i][4]) != float:
        continue
    if type(data_list[i+1][4]) != float:
        continue
    if data_list[i+1][4]-data_list[i][4] > change:
        change = data_list[i+1][4]-data_list[i][4]
change = round(change,2)
print('The largest price change between two days is '+ str(change))
```

```python
# 6. What was the average daily trading volume during this year?
total = 0
count = 0
for element in data_list:
    if type(element[6]) != float:
        continue
    total = total + element[6]
    count = count +1
average = total/count
average = round(average,2)
print('The average daily trading volume in 2017 is '+ str(average))
```

```python
# 7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)
trading_volumes = []
count= 0
for element in data_list:
    if type(element[6]) != float:
        continue
    trading_volumes.append(element[6])
    count = count + 1
trading_volumes.sort()

if count%2 == 0:
    median = (trading_volumes[int((count/2)-1)] + trading_volumes[int(count/2)])/2
else:
    median = trading_volumes[int((count-1)/2)]

print('The median daily trading volume in 2017 is '+ str(median))
```
